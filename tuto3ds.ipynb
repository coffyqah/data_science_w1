```python
#question 4
import numpy as np

arr = np.random.randint(1, 10, size = (6,3))

print("before transpose:")
print(arr, end = "\n\n")

print("after transpose:")
print(arr.transpose())

```

```python
#question 5
import numpy as np

arr1 = np.random.randint(1, 10, size = (2,3))
print("Array 1:")
print(arr1, end = "\n\n")
arr2 = np.random.randint(1, 10, size = (3,4))
print("Array 2:")
print(arr2, end = "\n\n")
print("Dot Product:")
print(np.dot(arr1, arr2))
```
